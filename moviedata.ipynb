## Answers for the following questions:

```python
import pandas as pd
```

The dataset is avaliable at https://www.kaggle.com/datasets/harshitshankhdhar/imdb-dataset-of-top-1000-movies-and-tv-shows?select=imdb_top_1000.csv

```python
data = pd.read_csv('imdb_top_1000.csv')
```

```python
print(data)
```

What are the column names in the dataset?

```python
for col in data.columns:
    print(col)
```

```python
ratings = data['IMDB_Rating'];
print(ratings)
```

What is the average rating of the movies?

```python
print(ratings.mean())
```

```python
genres = data['Genre'];
print(genres)
```

```python
print(genres.nunique())
```

How many unique genres are there?

```python
unique_genres = genres.value_counts()
```

```python
print(unique_genres)
```

Save your the results as a csv file

```python
unique_genres.to_csv('unique_genres.csv')
```

## Data Visualization

```python
from matplotlib import pyplot as plt
```

Plot a histogram of the genre distribution of the movies. How many movies from each genre are there in this list.

```python
plt.bar(unique_genres.index, unique_genres)
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Genre Distribution of Movies')
plt.xticks(rotation=90)
plt.show()
```

Plot a scatterplot in this dataset for the number of movies vs year (Each year, how many movies made to this top 1000 list?)

```python
movies = data.groupby('Released_Year').size()
```

```python
plt.scatter(movies.index, movies)
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies Released Each Year')
plt.show()
```
